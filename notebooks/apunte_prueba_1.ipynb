### Prueba de .pop()

El .pop() sin indicar la posicion devuelve el útimo elemento de la lista. Útil en algoritmos firts-in last-out

```python
lista_prueba = ['a', 'b', 'c', 'd', 'e']
poped_item = lista_prueba.pop()
print (lista_prueba, poped_item)
```

El .pop(0) devuelve el primer elemento de la lista. Útil en algoritmos firts-in first-out

```python
lista_prueba = ['a', 'b', 'c', 'd', 'e']
poped_item = lista_prueba.pop(0)
print (lista_prueba, poped_item)
```
