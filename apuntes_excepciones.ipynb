```python
# File mergedexc.py (Python 3.X + 2.X)
sep = '-' * 45 + '\n'

print(sep + 'EXCEPTION RAISED AND CAUGHT')

try:

    x = 'spam'[99]

except IndexError:  

    print('except run')

finally:

    print('finally run')
    
print('after run')
```

El programa intenta darle el valor `'spam'[99]` a `x`, como 99 está fuera de index salta un error que recoge el `except` y entonces ejecuta el `print()`
y después el `finally`.

```python
print(sep + 'NO EXCEPTION RAISED')

try:

    x = 'spam'[3]

except IndexError:

    print('except run')

finally:

    print('finally run')
    
print('after run')
```

En este caso no salta `IndexError`, por lo que ejecuta el codigo dentro del `try` y después el de `finally`.

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:

    x = 'spam'[3]

except IndexError:

    print('except run')

else:

    print('else run')

finally:
 
    print('finally run')
    
print('after run')
```

Misma situación que en el caso anterior, pero como nuevamente no se ejecuta el `except`, esta vez se ejecuta el codigo en el else, porque no se ejecuta ningún `except`.

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')

try:

    x = 1 / 0

except IndexError:

    print('except run')

finally:

    print('finally run')

print('after run')
```

El código está preparado para un `IndexError`, pero aquí está sucediendo una división entre 0.
Como el programa no está preparado para recivir el error, se detiene la ejecución.
`Finally` se ejecuta igualmente.
