## Objetos Inmutables

```python
a = 3
b = a               # b es 3
print("b es ", b)

a = 3
b = a
a = 'spam'          # b es 3
print("b es ", b)
```

'a' se iguala al valor 3, 'b' se iguala al valor de 'a', que previamente hemos dicho que es 3.

La idea es que b no se iguala a 'a' si no a 3, por lo tanto cuando variamos el valor de 'a', 'b' permanece inmutable.

a->3     
b->a->3 == b--->3

```python
a = 3
b = a
a = a + 2              # b es 3
print("b es ", b)
```

Ningún cambio que se realice sobre la variable 'a' se verá reflejado en 'b'.
"Los objetos **integer** no son mutables"

## OBJETOS MUTABLES
### Listas

```python
L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]

#-------------------------------------------

O = [2, 3, 4]
P = O			   
P[0] = 24 	

print("L = ", O)    # [24, 3, 4]
print("M = ", P)    # [24, 3, 4]
```

Al contrario que con los enteros, las variables que apuntan a listas apuntan todas a una misma lista, no a su valor.
Esto proboca que cualquier cambio realizado en la lista, se refleja en todas las variables que apunten a ella.

L->[lista]

M->L->[lista] != M--->[lista]

```python

L = [2, 3, 4]
M = L[:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed
```

De esta forma lo que copiamos no es la referancia a la lista, son los valores de esa lista en una nueva

### Shared References and Equality
- Operador `==` igualdad de valores.
- Operador `is` identidad de objetos.

```python

L = [1, 2, 3]
M = L 			    # M and L reference the same object

L == M 			    # True: Same values => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)
```

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M and L reference different objects

print(
    "L == M",
    L == M 			# True: Same values
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```

En el primer ejemplo ambas variables apuntan a la misma lista, por lo tanto sus valores son los mismo y además la propia lista también lo es.

En el segundo ejemplo cada variable referencia a una lista diferente, pero con los mismos valores.

### Cache

```python
X = 42
Y = 42		    # Should be two different objects

X  == Y         # True
X is Y 		    # True: Same object anyhow: caching at work!!
```

El cache es el valor interno al que apuntan los objetos. En el caso de los enteros no existen más de una copia de cada uno, pero puede haber varias variables apuntando al mismo valor.
En este ejemplo tanto el valor como la referencia son iguales en X e Y.

Para averiguar cuantas referencias tiene un objeto:

```python
import sys
sys.getrefcount(1)		# number of pointers to this shared piece of memory
```
